```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.model_selection import train_test_split as tts

import warnings
warnings.filterwarnings("ignore")
```

```python
iris_dataset = load_iris()
```

```python
trials = 50

# store accuracies for each n_neighbors value
avg_train_acc = []
avg_test_acc = []

for i in range(trials):
    X_train, X_test, y_train, y_test = tts(
        iris_dataset.data, iris_dataset.target, test_size=0.25, random_state=i)
    
    # store accuracies for each trials
    train_acc = []
    test_acc = []

    neighbors_settings = range(1, 41)

    for n_neighbors in neighbors_settings:
        clf = KNN(n_neighbors=n_neighbors)
        clf.fit(X_train, y_train)
        
        # train set accuracy
        train_acc.append(clf.score(X_train, y_train))
        
        # test set accuracy
        test_acc.append(clf.score(X_test, y_test))

    # average accuracies for trial
    avg_train_acc.append(train_acc)
    avg_test_acc.append(test_acc)

# average accuracies across all trials
avg_train_acc = np.mean(avg_train_acc, axis=0)
avg_test_acc = np.mean(avg_test_acc, axis=0)

# plotting
plt.figure(figsize=(9,6))
plt.plot(avg_train_acc, label="Train accuracy")
plt.plot(avg_test_acc, label="Test accuracy")
plt.xticks(ticks=range(len(neighbors_settings)), labels=list(neighbors_settings))
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.title("Mean Plot of Accuracy vs n_neighbors with 50 trials")
plt.legend()
plt.show()

print(f"Best Accuracy for KNN: {np.max(avg_test_acc):.4f}")
```

**Optimal n_neighbor** 

```python
# identify the optimal n_neighbors
n_neighbor = 13 - 1

plt.figure(figsize=(9,6))
plt.plot(avg_train_acc, label="Train accuracy")
plt.plot(avg_test_acc, label="Test accuracy")
plt.axvline(n_neighbor, label='Optimal n_neighbor', linestyle='--', linewidth=1, color='red')
plt.xticks(ticks=range(len(neighbors_settings)), labels=list(neighbors_settings))
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.title("Mean Plot of Accuracy vs n_neighbors with 50 trials")
plt.legend()
plt.show()

print(f"Best parameter for n_neighbor:", n_neighbor+1)
print(f"Train accuracy for the {n_neighbor+1}th neighbor: {avg_train_acc[n_neighbor]:.4f}")
print(f"Test accuracy for the {n_neighbor+1}th neighbor: {avg_test_acc[n_neighbor]:.4f}")
```

I labeled 13 as the optimal n_neighbor due to its notable attributes: it yields the highest test accuracy among all neighbors, while maintaining a reasonable difference to the training accuracy, indicating a balanced performance between generalization and overfitting.
